```python
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('/kaggle/input/uwjbfv/u.data', sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('/kaggle/input/uwjbfv/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

ratings = pd.merge(movies, ratings)
```

```python
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
```

```python
starWarsRatings = movieRatings['Star Wars (1977)']
starWarsRatings.head()
```

```python
similarMovies = movieRatings.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.sort_values(ascending=False)
```

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
popularMovies = movieStats['rating']['size'] >= 100

movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
mappedColumnsMoviestat=movieStats[popularMovies]
mappedColumnsMoviestat.columns=[f'{i}|{j}' if j != '' else f'{i}' for i,j in mappedColumnsMoviestat.columns]
df = mappedColumnsMoviestat.join(pd.DataFrame(similarMovies, columns=['similarity']))
```

```python
df.sort_values(['similarity'], ascending=False)[:15]
```

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
userRatings.head()
```

```python
corrMatrix=userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix=userRatings.corr(method='pearson' ,min_periods=100)
corrMatrix.head()
```

```python
myRatings=userRatings.loc[0].dropna()
myRatings.head()
```

```python
import pandas as pd

simCandidates = pd.Series([])  # Use pd.Series(dtype='float64') to avoid deprecation warning

for i in range(len(myRatings.index)):
    sims = corrMatrix[myRatings.index[i]].dropna()
    sims = sims.map(lambda x: x * myRatings[i])
    simCandidates = pd.concat([simCandidates, sims])

simCandidates.sort_values(ascending=False, inplace=True)
print(simCandidates.head(20))
```

```python
simCandidates=simCandidates.groupby(simCandidates.index).sum()
simCandidates.sort_values(ascending=False,inplace=True)
print(simCandidates.head(10))
```

```python
filteredCandidates = simCandidates.drop(labels=myRatings.index, errors='ignore')
print(filteredCandidates.head(10))
```
